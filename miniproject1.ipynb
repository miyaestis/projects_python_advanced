```python

class Neuron:
    def __init__(self, firing_rate=0):
        self.firing_rate = firing_rate

    def activate(self, stimulus_strength):
        """Calculate firing rate based on stimulus strength."""
        pass


class SensoryNeuron(Neuron):
    def __init__(self, firing_rate=0, receptor_type="Unknown"):
        super().__init__(firing_rate)
        self.receptor_type = receptor_type

    def sense_stimulus(self, stimulus):
        """Process stimulus and activate neuron."""
        pass


class MotorNeuron(Neuron):
    def __init__(self, firing_rate=0, target_muscle="Unknown"):
        super().__init__(firing_rate)
        self.target_muscle = target_muscle

    def control_muscle(self):
        """Trigger response in the target muscle."""
        pass


class Photoreceptor(SensoryNeuron):
    def __init__(self, firing_rate=0):
        super().__init__(firing_rate, receptor_type="light")

    def light_detection(self, light_intensity):
        """Activate based on light intensity."""
        pass


class Mechanoreceptor(SensoryNeuron):
    def __init__(self, firing_rate=0):
        super().__init__(firing_rate, receptor_type="pressure")

    def pressure_detection(self, pressure):
        """Activate based on pressure strength."""
        pass


class AlphaMotorNeuron(MotorNeuron):
    def __init__(self, firing_rate=0):
        super().__init__(firing_rate, target_muscle="skeletal muscle")

    def skeletal_muscle_control(self):
        """Initiate rapid contraction in skeletal muscle."""
        pass


class GammaMotorNeuron(MotorNeuron):
    def __init__(self, firing_rate=0):
        super().__init__(firing_rate, target_muscle="muscle spindle")

    def muscle_spindle_control(self):
        """Adjust muscle tone gradually."""
        pass

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


data = {
    "Company": ["Dell", "HP", "Apple", "Lenovo", "Asus", "Acer", "MSI"],
    "Price": [500, 600, 1200, 700, 550, 450, 1500],
    "RAM": [8, 16, 8, 16, 8, 4, 32],
    "Memory": ["256GB SSD", "512GB SSD", "256GB SSD", "1TB HDD", "256GB SSD", "128GB HDD", "1TB SSD"],
    "OpSys": ["Windows", "Windows", "MacOS", "Windows", "Windows", "Windows", "Windows"],
}

df = pd.DataFrame(data)


plt.figure(figsize=(10, 6))
plt.plot(df["Price"], 'o', label="Laptop Prices")
plt.title("Laptop Prices")
plt.xlabel("Laptop Index")
plt.ylabel("Price ($)")
plt.grid()
plt.legend()
plt.show()


avg_price_per_company = df.groupby("Company")["Price"].mean()
most_expensive_company = avg_price_per_company.idxmax()
print(f"The company with the most expensive laptops on average is {most_expensive_company}.")


df["OpSys"] = df["OpSys"].str.lower().replace({"windows": "Windows", "macos": "MacOS"})
unique_operating_systems = df["OpSys"].unique()
print(f"Unique operating systems: {unique_operating_systems}")


for os in unique_operating_systems:
    subset = df[df["OpSys"] == os]
    plt.figure(figsize=(10, 6))
    plt.hist(subset["Price"], bins=10, alpha=0.7, label=f"{os} Prices")
    plt.title(f"Price Distribution for {os}")
    plt.xlabel("Price ($)")
    plt.ylabel("Frequency")
    plt.legend()
    plt.grid()
    plt.show()


plt.figure(figsize=(10, 6))
plt.scatter(df["RAM"], df["Price"], alpha=0.7)
plt.title("Relationship Between RAM and Laptop Price")
plt.xlabel("RAM (GB)")
plt.ylabel("Price ($)")
plt.grid()
plt.show()


df["Storage Type"] = df["Memory"].apply(lambda x: "SSD" if "SSD" in x else "HDD")
print("Storage Types Count:")
print(df["Storage Type"].value_counts())

# Bonus question :)

plt.figure(figsize=(10, 6))
df["RAM"].value_counts().sort_index().plot(kind="bar", title="Most Common RAM Sizes")
plt.xlabel("RAM (GB)")
plt.ylabel("Frequency")
plt.grid()
plt.show()

# Example 2: Average price by storage type
average_price_by_storage = df.groupby("Storage Type")["Price"].mean()
plt.figure(figsize=(10, 6))
average_price_by_storage.plot(kind="bar", title="Average Price by Storage Type")
plt.xlabel("Storage Type")
plt.ylabel("Average Price ($)")
plt.grid()
plt.show()
```
